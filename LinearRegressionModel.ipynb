feature selection and visulization

```python
# Let's select some key features for our model
selected_features = ['RM', 'LSTAT', 'PTRATIO', 'INDUS']
X = df[selected_features]
y = df['MEDV']

# Visualize relationships
plt.figure(figsize=(15, 10))
for i, feature in enumerate(selected_features):
    plt.subplot(2, 2, i+1)
    plt.scatter(X[feature], y, alpha=0.5)
    plt.title(f'{feature} vs MEDV')
    plt.xlabel(feature)
    plt.ylabel('Median Value ($1000s)')
plt.tight_layout()
plt.show()
```

IMPLEMENTING LINEAR REGRESSION FROM SCRATCH

adding the bias term and feature scaling

```python
# FEATURE SCALING
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Add bias term (column of 1s)
X_b = np.c_[np.ones((len(X_scaled), 1)), X_scaled]

# Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X_b, y, test_size=0.2, random_state=42)
```

cost function implementation


```python
def compute_cost(X, y, theta):
    """
    Compute the MSE cost function

    Parameters:
    X : numpy array - Feature matrix with shape (m, n+1)
    y : numpy array - Target values with shape (m,)
    theta : numpy array - Parameters with shape (n+1,)

    Returns:
    float - Computed cost
    """
    m = len(y)
    predictions = X.dot(theta)
    cost = (1/(2*m)) * np.sum((predictions - y)**2)
    return cost
```

Gradient Descent Implementation

```python
def gradient_descent(X, y, theta, learning_rate=0.01, iterations=1000):
    """
    Perform gradient descent to learn theta

    Parameters:
    X : numpy array - Feature matrix
    y : numpy array - Target values
    theta : numpy array - Initial parameters
    learning_rate : float
    iterations : int - Number of iterations

    Returns:
    theta : numpy array - Learned parameters
    cost_history : list - Cost at each iteration
    """
    m = len(y)
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        gradient = (1/m) * X.T.dot(errors)
        theta -= learning_rate * gradient
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history
```

TRAINING THE MODEL

```python
# Initialize parameters
theta = np.random.randn(X_train.shape[1])

# Hyperparameters
learning_rate = 0.1
iterations = 1000

# Train the model
theta, cost_history = gradient_descent(X_train, y_train, theta, learning_rate, iterations)

# Print final parameters
print("Learned parameters:")
for i, feature in enumerate(['Bias'] + selected_features):
    print(f"{feature}: {theta[i]:.4f}")
```

EVALUATION THE MODEL

PLOTTING TRAINNING PROGRESS

```python
# Plot cost over iterations
plt.figure(figsize=(10, 6))
plt.plot(range(iterations), cost_history)
plt.title('Cost Function Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.grid(True)
plt.show()
```

MAKING PREDICTION AND CALCULATION METRICES

```python
def predict(X,theta):
  """ Make prediction used learned theta """
  return X.dot(theta)
# make predictions
y_pred=predict(X_test,theta)

# calculate metrices
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"R-squared Score: {r2:.2f}")
```

Comparing with Scikit-Learn's Implementation

```python
from sklearn.linear_model import LinearRegression

# Create and fit sklearn's linear regression
lin_reg = LinearRegression()
lin_reg.fit(X_train[:, 1:], y_train)  # Exclude bias term as sklearn adds it automatically

# Compare parameters
print("\nComparison with sklearn:")
print("Our intercept:", theta[0])
print("Sklearn intercept:", lin_reg.intercept_)
print("\nOur coefficients:", theta[1:])
print("Sklearn coefficients:", lin_reg.coef_)

# Compare predictions
sklearn_pred = lin_reg.predict(X_test[:, 1:])
print(f"\nPrediction MSE difference: {mean_squared_error(y_pred, sklearn_pred):.5f}")
```
